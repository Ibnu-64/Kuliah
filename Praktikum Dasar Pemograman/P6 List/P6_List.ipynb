## 1. **Pendahuluan**

List adalah salah satu struktur data yang paling sering digunakan dalam Python karena fleksibilitasnya. List dapat menyimpan berbagai jenis data dan memungkinkan kita melakukan berbagai operasi seperti menambah, menghapus, dan mengubah elemen.

## 2. **Pembahasan Singkat Fungsi dan Method pada List**

| **No** | **Fungsi/Method**   | **Penjelasan**                                                                                   | **Contoh Kode**                                                                                                                                  |
|--------|---------------------|---------------------------------------------------------------------------------------------------|--------------------------------------------------------------------------------------------------------------------------------------------------|
| 1      | `len()`              | Menghitung jumlah elemen dalam List.                                                             | ```python len([1, 2, 3])  # Output: 3```                                                                                                         |
| 2      | `min()`              | Mengembalikan elemen dengan nilai terkecil (hanya bisa digunakan pada elemen yang bisa dibandingkan). | ```python min([4, 2, 8, 1])  # Output: 1```                                                                                                     |
| 3      | `max()`              | Mengembalikan elemen dengan nilai terbesar.                                                      | ```python max([4, 2, 8, 1])  # Output: 8```                                                                                                     |
| 4      | `append()`           | Menambahkan elemen baru di akhir List.                                                          | ```python buah = ['apel', 'jeruk'] buah.append('pisang')  # Output: ['apel', 'jeruk', 'pisang']```                                               |
| 5      | `insert()`           | Menyisipkan elemen baru pada posisi tertentu dalam List.                                         | ```python buah = ['apel', 'jeruk'] buah.insert(1, 'mangga')  # Output: ['apel', 'mangga', 'jeruk']```                                           |
| 6      | `extend()`           | Menggabungkan dua List atau koleksi data lainnya ke akhir List.                                  | ```python buah = ['apel'] sayur = ['bayam'] buah.extend(sayur)  # Output: ['apel', 'bayam']```                                                   |
| 7      | `remove()`           | Menghapus elemen pertama yang ditemukan dengan nilai tertentu dari List.                        | ```python buah = ['apel', 'jeruk', 'mangga'] buah.remove('jeruk')  # Output: ['apel', 'mangga']```                                               |
| 8      | `pop()`              | Menghapus elemen pada indeks tertentu (default: elemen terakhir) dan mengembalikannya.           | ```python buah = ['apel', 'jeruk', 'mangga'] buah.pop()  # Output: 'mangga'```                                                                  |
| 9      | `clear()`            | Menghapus semua elemen dalam List.                                                              | ```python buah = ['apel', 'jeruk'] buah.clear()  # Output: []```                                                                                 |
| 10     | `sort()`             | Mengurutkan elemen dalam List secara menaik (secara default).                                   | ```python angka = [5, 2, 9] angka.sort()  # Output: [2, 5, 9]```                                                                                 |
| 11     | `reverse()`          | Membalik urutan elemen dalam List.                                                              | ```python angka = [1, 2, 3] angka.reverse()  # Output: [3, 2, 1]```                                                                             |
| 12     | `copy()`             | Membuat salinan List baru dari List yang ada.                                                   | ```python buah = ['apel'] buah_copy = buah.copy()  # Output: ['apel']```                                                                         |
| 13     | `count()`            | Menghitung jumlah kemunculan elemen tertentu dalam List.                                        | ```python buah = ['apel', 'apel', 'jeruk'] buah.count('apel')  # Output: 2```                                                                   |
| 14     | `index()`            | Mengembalikan indeks dari elemen pertama yang ditemukan.                                         | ```python buah = ['apel', 'jeruk'] buah.index('jeruk')  # Output: 1```                                                                          |

### **Soal 1 - Fungsi `len()`, `min()`, dan `max()`**
<img src="Screenshot (1).png" alt="Deskripsi Gambar" width="450" style="margin-left: 30px;" />

**Instruksi:**
1. Buat sebuah List berisi angka NIM dibagi 2 value tiap NIM nya.
2. Tampilkan jumlah elemen dalam List, nilai terkecil, dan nilai terbesar.
3. Jelaskan code beserta outputnya!

````python
# Ketikan Code Disini lalu Jalankan (Run)

#  List dari nim 2411102441264
list_nim = [24,11,10,24,41,26,4]

panjang_list = len(list_nim) # ambil panjang list
min_list = min(list_nim) # ambil nilai terkecil dari list
max_list = max(list_nim) # anim nilai terbesar dari list

# cetak hasil
print(f"Jumlah elemen: {panjang_list}")
print(f"Nilai terkecil dari list nim: {min_list}")
print(f"Nilai terbesar dari list nim: {max_list}")
````

### **Soal 2 - Method `append()`, `insert()`, dan `extend()`**

1. Buatlah List berisi nama-nama kota.

### **Soal 3 - Method `pop()`, `remove()`, dan `clear()`**
<img src="Screenshot (3).png" alt="Deskripsi Gambar" width="450" style="margin-left: 30px;" />

**Instruksi:**
1. Buat sebuah List berisi nama buah.
2. Hapus elemen terakhir dari List menggunakan `pop()`.
3. Hapus buah tertentu dari List menggunakan `remove()`.
4. Kosongkan seluruh List menggunakan `clear()`.

````python
# Ketikan Code Disini lalu Jalankan (Run)
# List buah
buah = ['Apel', 'Jeruk', 'Mangga', 'Durian', 'Salak', 'Nanas', 'Pisang', 'Anggur', 'Melon', 'Semangka']
#hapus elemen terakhir dari list buah
buah.pop()
print(f"list buat setelah elemen terakhir di hapus: {buah}")

#hapus elemen "durian" dari list buah
buah.remove('Durian')
print(f"list buat setelah durian di hapus: {buah}")

#hapus semua elemen dari list buah
buah.clear()
print(f"list buat setelah semua elemen di hapus: {buah}")

````

### **Soal 4 - Method `sort()` dan `reverse()`**
<img src="Screenshot (4).png" alt="Deskripsi Gambar" width="300" style="margin-left: 30px;" />

**Instruksi:**
1. Buatlah sebuah List berisi beberapa angka.
2. Urutkan List tersebut dalam urutan menaik menggunakan `sort()`.
3. Balik urutan List menggunakan `reverse()`.

````python
# Ketikan Code Disini lalu Jalankan (Run)

 # List nums
nums = [3,63,45,23,73,95,2,5,3,7,9,9,32,44,55,66,77,88,99,100]

# mengurutkan list nums
nums.sort()
print(f"nums yang sudah di ururtkan: {nums}")# cetak hasil

# membalik urutan elemen dari dari nums yang sudah di urutkan
nums.reverse() 
print(f"nums yang telah di urutan dan di balik: {nums}") # cetak hasil

#output = nums yang sudah di ururtkan: [2, 3, 3, 5, 7, 9, 9, 23, 32, 44, 45, 55, 63, 66, 73, 77, 88, 95, 99, 100]
#         nums yang telah di urutan dan di balik: [100, 99, 95, 88, 77, 73, 66, 63, 55, 45, 44, 32, 23, 9, 9, 7, 5, 3, 3, 2]

# Karena list elemen nums akan di ururtkan dari yang terbesar ke yang terkecil, setelah di urutkan akan di balik urutannya
````

### **Soal 5 - Menerapkan Semua Fungsi dan Method**
<img src="Screenshot (5).png" alt="Deskripsi Gambar" width="450" style="margin-left: 30px;" />

**Instruksi:**
1. Buatlah program yang meminta pengguna memasukkan 5 nilai numerik.
2. Simpan nilai-nilai tersebut ke dalam List.
3. Gunakan:
   - `len()` untuk menghitung jumlah elemen.
   - `min()` dan `max()` untuk mencari nilai terkecil dan terbesar.
   - `append()` untuk menambahkan nilai baru yang diminta dari pengguna.
   - `pop()` untuk menghapus elemen terakhir.
   - `sort()` untuk mengurutkan nilai-nilai dalam List.
4. Tampilkan hasil setelah setiap operasi dan jelaskan code beserta outputnya!.



````python
# Ketikan Code Disini lalu Jalankan (Run)

nums = [] # List kosong untuk menampung angka yang di inputkan

# meminta user untuk memasukkan 5 angka
for i in range(5):
    num = int(input(f"Masukkan angka ke-{i+1}: ")) # meminta user untuk memasukkan angka
    nums.append(num) # tambahkan angka yang di inputkan ke list nums

print(f"List angka: {nums}") # cetak hasil nums

new_num = int(input("Masukkan angka baru: ")) # meminta user untuk memasukkan angka baru
nums.append(new_num) # tambahkan angka baru ke belakang list nums


print(f"Setelah menambah angka: {nums}") # cetak hasil setelah menambah angka

nums.pop() # hapus elemen terakhir dari list nums
print(f"Setelah menghapus angka terakhir: {nums}") # cetak hasi setelah menghapus angka terakhir

nums.sort() # mengurutkan list nums
print(f"Setelah nums diurutkan: {nums}") # cetak hasil setelah


print(f"Nilai terkecil dari nums: {min(nums)}") # ambil nilai terkecil dari list nums dan cetak hasil

print(f"Nilai terbesar dari nums: {max(nums)}") # ambil nilai terbesar dari list nums dan cetak hasil

print(f"Panjang dari nums: {len(nums)}") # ambil panjang list nums dan cetak hasil cetak hasil


#penjelasan ouput
# nums akan di manipulasi seteap metode yang di jalankan


````

1. Buatlah program untuk mengelola daftar belanja menggunakan List, di mana pengguna bisa menambahkan, menghapus, dan melihat item dalam daftar belanja tersebut.
2. Implementasikan semua fungsi dan method yang telah dipelajari pada skenario yang lebih relevan dengan aplikasi dunia nyata.


````python
daftar_belanja = {}; # Menggunakan dictionary untuk item dan jumlahnya

# Fungsi untuk mencetak opsi
def print_opsi():
    print("\n--- Pilih opsi ---")
    print("1. Tambah item ke daftar belanja")
    print("2. Hapus item dari daftar belanja")
    print("3. Lihat daftar belanja")
    print("4. Jumlah belanja")
    print("5. Keluar")
    print("------------------")

# Fungsi untuk menambahkan item ke daftar belanja
def addItem(item):
    if item in daftar_belanja:  
        daftar_belanja[item] += 1 # Jika item sudah ada, tambahkan jumlahnya
    else:
        daftar_belanja[item] = 1 # Jika item belum ada, tambahkan item tersebut
    
    print(f'"{item}" telah ditambahkan ke daftar belanja!')

# Fungsi untuk menghapus item dari daftar belanja
def removeItem(item):
    # Periksa apakah item ada di daftar belanja
    if item in daftar_belanja:
        if daftar_belanja[item] > 1:
            daftar_belanja[item] -= 1 # Jika jumlah item lebih dari 1, kurangi jumlahnya
            print(f'jumlah "{item}" dikurangi. (Sisa item: {daftar_belanja[item]}))')
        else:
            del daftar_belanja[item] # Jika jumlah item hanya 1, hapus item tersebut (del adalah keyword untuk menghapus item dari list beserta kunci dan nilainya)
            print(f'"{item}" telah di hapus dalam daftar belanja!')
    else:
        print(f'"{item}" tidak ada di daftar belanja!')
# Fungsi untuk menampilkan daftar belanja
def showList():
    if not daftar_belanja:  # Jika daftar belanja kosong
        print("\nDaftar belanja kosong.")
    else:
        print("\nDaftar belanja:")

        # ambil item dan jumlahnya dari daftar belanja
        for item, jumlahItem in daftar_belanja.items():
            print(f"* {item} ({jumlahItem})") # Cetak item dan jumlahnya



# Looping untuk meminta user memilih opsi
while True:
    try:
        print_opsi()  # Cetak opsi
        # Meminta user untuk memasukkan angka
        opsi = int(input("Pilih Opsi: "))

        # jika user memasukkan angka yang tidak valid
        if opsi < 1 or opsi > 5:
            raise ValueError("Angka harus di antara 1 dan 5!")
        
    except ValueError as error:
        # Jika user memasukkan bukan 4angka
        print(f"Error: {error}")
    except KeyboardInterrupt:
        # Jika pengguna menekan Ctrl+C
        print("\nInput dihentikan oleh pengguna.")
        break
    else: 

        # user memilih tambah item ke daftar belanja
        if opsi == 1:
            item = input("Apa yang ingin anda tambahkan ke daftar belanja :").lower() # Meminta user untuk memasukkan item
            addItem(item) # Menambahkan item ke daftar belanja

        #user memilih hapus item dari daftar belanja
        elif opsi == 2:
            item = input("item apa yang ingin anda hapus :").lower() # Meminta user untuk memasukkan item
            removeItem(item) # Menghapus item dari daftar belanja

        #user memilih melihat daftar belanja
        elif opsi == 3:
            showList()     
        # user memilih keluar dari program
        elif opsi == 4:
            print(f"Jumlah belanjan anda sekarang andalah: {sum(daftar_belanja.values())}") # Menghitung jumlah belanja
    
        elif opsi == 5:
            print("Anda telah keluar dari program.")
            break

    

````

# Kesimpulan

List dalam Python merupakan struktur data sering digunakan karena kemampuannya untuk menyimpan berbagai jenis data dan melakukan berbagai manipulasi. Penggunaan yang tepat dari fungsi dan method ini akan memudahkan dlm pengolahan data. Dengan memahami dan menguasai List, kita dpt menerapkan logika pemrograman untuk menyelesaikan berbagai masalah yang berkaitan dengan pengolahan data.

Beberapa fungsi dan method yang sering digunakan dengan List antara lain:
1. Penghitungan dan Perbandingan: Fungsi len(), min(), dan max() memungkinkan pengguna untuk menghitung jumlah elemen serta menemukan nilai terkecil dan terbesar dalam List.

2. Manipulasi Data: Method seperti append(), insert(), dan extend() digunakan untuk menambahkan elemen ke dalam List. Sebaliknya, remove(), pop(), dan clear() digunakan untuk menghapus elemen dari List.

3. Pengaturan Urutan: Method sort() dan reverse() membantu dalam mengurutkan dan membalikkan elemen dalam List.
